# Welcome to Jupyter!

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
user_data=pd.read_csv('user_data.csv')
user_data
```

```python
logs=pd.read_csv('logs.csv')
logs
```

```python
user_data.describe()
```

```python
# посчитать уникальные
logs.platform.nunique()
```

```python
# клиенты с максимальным числом успешных действий
max_suc = logs.query('success == True')\
              .groupby('client', as_index=False)\
              .agg({'success':'count'}) \
              .sort_values(['success', 'client'], ascending=[False, True]).query('success == 41')
```

```python
l = max_suc.client.tolist()
l
```

```python
# получить строку из списка
', '.join([str(i) for i in l])
```

```python
logs.groupby('platform', as_index=False).agg({'success':'sum'})
```

```python
join_data=logs.merge(user_data, on='client')
join_data
```

```python
#Предпочтения премиальных клиентов по платформам
join_data.groupby('platform', as_index=False) \
         .agg({'premium':'sum'}) \
         .sort_values('premium', ascending=False)
```

```python
# распределение возраста клиентов в зависимости от типа клиента (премиум или нет)

sns.distplot(join_data.query('premium == False').age)
sns.distplot(join_data.query('premium == True').age)
```
